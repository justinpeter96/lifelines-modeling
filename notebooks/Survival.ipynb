```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import lifelines
from lifelines import KaplanMeierFitter
from lifelines import NelsonAalenFitter
from lifelines.utils import median_survival_times
from lifelines.statistics import logrank_test
from lifelines import CoxPHFitter
from sklearn.model_selection import train_test_split
```

```python
mean = pd.read_csv("../data/raw/mean.csv",index_col=[0])
median = pd.read_csv("../data/raw/median.csv",index_col=[0])
mode = pd.read_csv("../data/raw/mode.csv",index_col=[0])
iterative = pd.read_csv("../data/raw/iterative.csv",index_col=[0])
```

```python
mean['Gender'].replace([0.0,1.0],['F','M'],inplace=True)
median['Gender'].replace([0.0,1.0],['F','M'],inplace=True)
mode['Gender'].replace([0.0,1.0],['F','M'],inplace=True)
iterative['Gender'].replace([0.0,1.0],['F','M'],inplace=True)
```

```python
kmf = KaplanMeierFitter() # 'Haemoglobin','Albumin',
x1 = mean['Iron'] 
y1 = mean['Class']
x2 = median['Iron'] 
y2 = median['Class']
x3 = mode['Iron'] 
y3 = mode['Class']
x4 = iterative['Iron'] 
y4 = iterative['Class']
```

```python
kmf1 = kmf.fit(x1,y1)
kmf2 = kmf.fit(x2,y2)
kmf3 = kmf.fit(x3,y3)
kmf4 = kmf.fit(x4,y4)
```

```python
kmf1.survival_function_.plot()
plt.title('Mean Survival based on Iron');
```

```python
kmf2.survival_function_.plot()
plt.title('Median Survival based on Iron');
```

```python
kmf3.survival_function_.plot()
plt.title('Most Frequent Survival based on Iron');
```

```python
kmf4.survival_function_.plot()
plt.title('Iterative Survival based on Iron');
```

```python
kmfn = [kmf1,kmf2,kmf3,kmf4]
ci = []
for i in kmfn:
    print(i.median_survival_time_)
    b = median_survival_times(i.confidence_interval_)
    ci.append(b)
print(ci)
```

### Survival Regression 

```python
mean = pd.read_csv("../data/raw/mean.csv",index_col=[0])
median = pd.read_csv("../data/raw/median.csv",index_col=[0])
mode = pd.read_csv("../data/raw/mode.csv",index_col=[0])
iterative = pd.read_csv("../data/raw/iterative.csv",index_col=[0])
```

```python
### helper functions

def get_data(data_name):
    
    if data_name == 'median':
        data = median
    elif data_name == 'mean':
        data = mean
    elif data_name == 'mode':
        data = mode
    else:
         data = iterative
    
    X = data.drop(['Class'], axis=1) # get independent variable
    y = data['Class'] # get dependent variable
    
    # split data 70% to 30%
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)
    
    return X_train, X_test, y_train, y_test
```

```python
columns = ["mean", "median", "mode", "iterative"]

for i in columns:
    
    print('\n')
    print(i.upper(), ' IMPUTED DATASET')
    
    cph = CoxPHFitter()
    
    X_train, X_test, y_train, y_test = get_data("mean")

    data = pd.concat([X_train, y_train], axis=1)
    data = data.reset_index()
    
    cph_fit = cph.fit(data[['Haemoglobin', 'Albumin', 'Iron', 'Age at diagnosis', 'Class']], duration_col='Age at diagnosis', event_col='Class')
    cph_fit.print_summary()
    
    cph.plot()
    plt.show()
```
